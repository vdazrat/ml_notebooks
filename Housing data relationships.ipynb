Training the sets, split the data into test and train sets. This can be done randomly using the sklearn cross-validation

```python
X_train,X_test,Y_train,Y_test = sklearn.model_selection.train_test_split(X,bos.price,test_size=0.33,random_state=5)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
lm = LinearRegression()
lm.fit(X_train,Y_train)
pred_train = lm.predict(X_train)
pred_test = lm.predict(X_test)
```

```python
print("RME for the training data",np.mean((Y_train - lm.predict(X_train))**2))
print("RME for the test data",np.mean((Y_test - lm.predict(X_test))*2))
```

```python
plt.scatter(Y_test,lm.predict(X_test))
plt.xlabel("actual test data price")
plt.ylabel("predictions for the test data")
```

Residual Plots - how far off is the predicted value from the actual. How big is the residual for the train vs test

```python
plt.scatter(lm.predict(X_train), lm.predict(X_train) - Y_train, c='b', s=40, alpha=0.5)
plt.scatter(lm.predict(X_test), lm.predict(X_test) - Y_test, c='g', s=40)
plt.hlines(y = 0, xmin=0, xmax = 50)
plt.title('Residual Plot using training (blue) and test (green) data')
plt.ylabel('Residuals')
```

```python
X = faithful.waiting
y = faithful.eruptions
model = sm.OLS(y, X)
results = model.fit()
```

```python
results.summary()
```

```python
results.params.values
```

```python
X = sm.add_constant(X)
X.head()
```

```python
modelW0 = sm.OLS(y, X)
resultsW0 = modelW0.fit()
resultsW0.summary()
```

```python
newX = np.array([1,75])
```

```python
resultsW0.params[0]*newX[0] + resultsW0.params[1] * newX[1]
```
